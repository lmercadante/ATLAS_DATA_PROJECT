```python
MJD_DAT_13513_g = []
MJD_SIM_13513_g = []
DATA_13513_g = []
SIM_13513_g = []

MJD_DAT_13513_r = []
MJD_SIM_13513_r = []
DATA_13513_r = []
SIM_13513_r = []

MJD_DAT_13513_i = []
MJD_SIM_13513_i = []
DATA_13513_i = []
SIM_13513_i = []

for index, row in df.iterrows():
    ID = row[1]
    MJD = row[2]
    FLUX = row[5]
    DATAFLAG = row[7]
    FILTER = row[8]
    if ID == 13513:
        if FILTER == 'g':
            if DATAFLAG == 1:
                DATA_13513_g.append(FLUX)
                print(FLUX)
                MJD_DAT_13513_g.append(MJD)
            elif DATAFLAG == 0:
                SIM_13513_g.append(FLUX) 
                MJD_SIM_13513_g.append(MJD)
        elif FILTER == 'r':
            if DATAFLAG == 1:
                DATA_13513_r.append(FLUX)
                MJD_DAT_13513_r.append(MJD)
            elif DATAFLAG == 0:
                SIM_13513_r.append(FLUX) 
                MJD_SIM_13513_r.append(MJD)
        elif FILTER == 'i':
            if DATAFLAG == 1:
                DATA_13513_i.append(FLUX)
                MJD_DAT_13513_i.append(MJD)
            elif DATAFLAG == 0:
                SIM_13513_i.append(FLUX) 
                MJD_SIM_13513_i.append(MJD)
    else:
        print('Not Data or Sim')
```

```python
len(DATA_13513)
plt.plot(MJD_DAT_13513_g,DATA_13513_g, 'g.')
plt.plot(MJD_SIM_13513_g,SIM_13513_g, color = 'green')

plt.plot(MJD_DAT_13513_r,DATA_13513_r, 'r.')
plt.plot(MJD_SIM_13513_r,SIM_13513_r, color = 'red')

plt.plot(MJD_DAT_13513_i,DATA_13513_i, 'b.')
plt.plot(MJD_SIM_13513_i,SIM_13513_i, color = 'blue')
```

```python
ID_list = np.unique(df['CID'])
```

```python
for i in ID_list:
    MJD_DAT_g = []
    MJD_SIM_g = []
    DATA_g = []
    SIM_g = []

    MJD_DAT_r = []
    MJD_SIM_r = []
    DATA_r = []
    SIM_r = []

    MJD_DAT_i = []
    MJD_SIM_i = []
    DATA_i = []
    SIM_i = []
    
    for index, row in df.iterrows():
        ID = row[1]
        MJD = row[2]
        FLUX = row[5]
        DATAFLAG = row[7]
        FILTER = row[8]
        if ID == i:
            if FILTER == 'g':
                if DATAFLAG == 1:
                    DATA_g.append(FLUX)
                    MJD_DAT_g.append(MJD)
                elif DATAFLAG == 0:
                    SIM_g.append(FLUX) 
                    MJD_SIM_g.append(MJD)
            elif FILTER == 'r':
                if DATAFLAG == 1:
                    DATA_r.append(FLUX)
                    MJD_DAT_r.append(MJD)
                elif DATAFLAG == 0:
                    SIM_r.append(FLUX) 
                    MJD_SIM_r.append(MJD)
            elif FILTER == 'i':
                if DATAFLAG == 1:
                    DATA_i.append(FLUX)
                    MJD_DAT_i.append(MJD)
                elif DATAFLAG == 0:
                    SIM_i.append(FLUX) 
                    MJD_SIM_i.append(MJD)
            
        
    plt.figure()
    plt.xlabel('MJD', size = 15)
    plt.ylabel('Flux', size = 15)
    plt.title('Lightcurve for SN' + str(i), size = 18)
    plt.plot(MJD_DAT_g,DATA_g, 'g.')
    plt.plot(MJD_SIM_g,SIM_g, color = 'green', label = 'g')

    plt.plot(MJD_DAT_r,DATA_r, 'r.')
    plt.plot(MJD_SIM_r,SIM_r, color = 'red', label = 'r')

    plt.plot(MJD_DAT_i,DATA_i, 'b.')
    plt.plot(MJD_SIM_i,SIM_i, color = 'blue', label = 'i')
    plt.legend(loc = 'upper right')
```
