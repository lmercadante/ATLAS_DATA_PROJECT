```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
df_files = pd.read_csv('/project2/rkessler/SURVEYS/LSST/USERS/lmercadante/ATLAS_DATA_PROJECT/ATLAS_API_OUTPUT/CLEAN_DATA/file_list_c.csv', sep = ',')
file_list = df_files['File_Name']
#df = pd.read_csv('/project2/rkessler/SURVEYS/LSST/USERS/lmercadante/ATLAS_DATA_PROJECT/ATLAS_API_OUTPUT/RAW_DATA/file_list.csv', sep = ',')
name_list = df_files['SN_Name']

path = '/project2/rkessler/SURVEYS/LSST/USERS/lmercadante/ATLAS_DATA_PROJECT/ATLAS_API_OUTPUT/CLEAN_DATA/'
```

```python
#df_files
```

```python
len(file_list)
```

```python
for i in range(0,len(file_list)):
    df = pd.read_csv(path+file_list[i], sep=",")

    MAG_AB_C = []
    MAG_ERR_C = []
    MJD_C = []
    MAG_AB_O = []
    MAG_ERR_O = []
    MJD_O = []

    for index, row in df.iterrows():
        MJD = row[1]
        MAG_AB = row[2]
        MAG_ERR = row[3]
        FILTER = row[6]
        if FILTER == 'c':
            if MAG_AB < 0 or MAG_ERR > 1:
                print("Outlier")
            else:
                MAG_AB_C.append(MAG_AB)
                MAG_ERR_C.append(MAG_ERR)
                MJD_C.append(MJD)
        else:
            if MAG_AB < 0 or MAG_ERR > 1:
                print("Outlier")
            else:
                MAG_AB_O.append(MAG_AB)
                MAG_ERR_O.append(MAG_ERR)
                MJD_O.append(MJD)
        
    fig, ax = plt.subplots()
    ax.errorbar(MJD_C, MAG_AB_C, ecolor = 'c', yerr= MAG_ERR_C,fmt='c.')
    ax.errorbar(MJD_O, MAG_AB_O, ecolor = 'orange', yerr= MAG_ERR_O,fmt='.', color = 'orange')
    ax.set_ylim([23,15])
    ax.set(ylabel = 'AB Magnitude')
    ax.set(xlabel = 'MJD')
    plt.title('Cleaned Data for '+str(name_list[i]))
    #time.sleep(15)
```

```python
fig, ax = plt.subplots()
ax.errorbar(MJD_C, MAG_AB_C, ecolor = 'c', yerr= MAG_ERR_C,fmt='c.')
ax.errorbar(MJD_O, MAG_AB_O, ecolor = 'orange', yerr= MAG_ERR_O,fmt='.', color = 'orange')
```
